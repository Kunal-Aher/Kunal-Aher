```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

def extract_table_from_url(url):
    """Extracts the specific table from a given URL."""
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise HTTPError for bad responses

        soup = BeautifulSoup(response.content, 'html.parser')
        table = soup.find('table', class_='views-table views-view-table cols-5')

        if table:
            data = []
            for row in table.find('tbody').find_all('tr'):
                cells = row.find_all('td')
                if len(cells) == 5:
                    try:
                        logo_cell = cells[0].find('a')
                        logo_link = logo_cell['href'] if logo_cell and 'href' in logo_cell.attrs else None
                        logo_image = logo_cell.find('img')['src'] if logo_cell and logo_cell.find('img') and 'src' in logo_cell.find('img').attrs else None

                        investor_name_element = cells[1].find('a')
                        investor_name = investor_name_element.text.strip() if investor_name_element else cells[1].text.strip() if cells[1].text else None

                        country = cells[2].text.strip() if cells[2].text else None
                        profile = cells[3].text.strip() if cells[3].text else None
                        deals = cells[4].text.strip() if cells[4].text else None

                        logo_link_full = f"https://kando.tech/{logo_link}" if logo_link else None
                        logo_image_full = f"https://kando.tech/{logo_image}" if logo_image else None

                        data.append({
                            'Logo Link': logo_link_full,
                            'Logo Image': logo_image_full,
                            'Investor': investor_name,
                            'Country': country,
                            'Profile': profile,
                            'Deals': deals
                        })
                    except Exception as e:
                        print(f"Error processing row: {e}")
            return data
        else:
            print(f"Table not found on {url}")
            return None
    except requests.exceptions.HTTPError as e:  #Specific handling for HTTP errors (like 500)
        print(f"HTTPError for {url}: {e}")
        return None
    except requests.exceptions.RequestException as e: #Catch all other request exceptions
        print(f"Request Exception for {url}: {e}")
        return None
    except Exception as e:
        print(f"General Exception for {url}: {e}")
        return None


# 1. Read the Excel file containing the website links
excel_file = "pagination_links.xlsx"  # Replace with the actual filename
try:
    df_links = pd.read_excel(excel_file,sheet_name='Sheet5')
except FileNotFoundError:
    print(f"Error: File '{excel_file}' not found.")
    exit()
except Exception as e:
    print(f"Error reading Excel file: {e}")
    exit()

# Assuming the column containing the website links is named 'website_link'
if 'Pagination Links' not in df_links.columns:
    print("Error: 'Pagination Links' column not found in the Excel file.")
    exit()

all_tables_data = []  # a list to hold ALL scraped rows from ALL sites

# 2. Iterate through the website links and extract the table data
start_time = time.time()  # Start timer

for index, row in df_links.iterrows():
    website_url = row['Pagination Links']  # Fetch URL from the 'website_link' column
    print(f"Processing URL: {website_url}")

    #Retry Logic
    max_retries = 3
    retry_delay = 5  # seconds
    for attempt in range(max_retries):
        try:
            table_data = extract_table_from_url(website_url)
            if table_data:
                all_tables_data.extend(table_data)  # Add the extracted data to the combined list
            break  # Success, exit the retry loop

        except Exception as e:
            print(f"Attempt {attempt + 1} failed: {e}")
            if attempt < max_retries - 1:
                time.sleep(retry_delay)
            else:
                print(f"Max retries reached for {website_url}. Skipping.")

    time.sleep(1)  # Add delay to avoid rate limiting

# 3. Create a Pandas DataFrame from all the extracted data
df_combined = pd.DataFrame(all_tables_data)

# 4. Export the combined DataFrame to a new Excel file
output_excel_file = "ContinuumMedia_Round2_assignment.xlsx"
df_combined.to_excel(output_excel_file, index=False)

end_time = time.time()
total_time = end_time - start_time

print(f"Extraction and combination completed in {total_time:.2f} seconds.")
print(f"Combined data saved to {output_excel_file}")
```
