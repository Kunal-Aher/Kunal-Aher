```python
#logistic regression
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

```python
Task1=pd.read_excel(r"D:\kunal_cdac_blr_notes\ML\TASK_ONE.xlsx")
```

```python
Task1
```

```python
Task1.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
Task1
```

```python
Task1['placement']=Task1['placement'].map({"placed":1,"not placed":0})
```

```python
Task1
```

```python
sns.heatmap(Task1.corr(),annot=True)
```

```python
x=Task1[['cgpa','study_hrs','screen_hrs']]
x1=Task1[['study_hrs']]
y=Task1['placement']
```

```python
x1_train,x1_test,y_train,y_test=train_test_split(x1,y,test_size=0.30,random_state=42)
```

```python
model=LogisticRegression()
model.fit(x1_train,y_train)
```

```python
coeff=model.coef_
intercept=model.intercept_
```

```python
line_eq=coeff*10+intercept
```

```python
line_eq
```

```python
y_pred=model.predict(x1_test)
```

```python
y_pred
```

```python
data=pd.DataFrame({"predicted":y_pred,"Actual":y_test})
data
```

```python
confu_matrix=confusion_matrix(y_test,y_pred)
```

```python
confu_matrix
```

```python
class_report=classification_report(y_test,y_pred)
```
