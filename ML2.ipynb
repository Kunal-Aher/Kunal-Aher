```python
Task9:-
=======
what is NORMAL DISTRIBUTION ?

how to find the skewness without a graph ?

how to identify the left skewed / right skewed ?

How to find the kurtosis without a graph ?





Great questions! Let's dive into each one:

### 1. What is Normal Distribution?

The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetrical around its mean. In a graph, it forms a bell-shaped curve where the highest point corresponds to the mean. Here are some key properties:
- **Symmetrical**: The left and right sides of the curve are mirror images.
- **Mean, Median, and Mode**: All three are located at the center.
- **Empirical Rule**: About 68% of the data falls within one standard deviation from the mean, 95% within two, and 99.7% within three.

### 2. How to Find the Skewness Without a Graph?

Skewness quantifies the asymmetry of a distribution. Heres how you can calculate it without plotting a graph:
- **Formula**: 
  \[
  \text{Skewness} = \frac{n}{(n-1)(n-2)} \sum \left(\frac{x_i - \overline{x}}{s}\right)^3
  \]
  where \( n \) is the number of data points, \( x_i \) are the data points, \( \overline{x} \) is the mean, and \( s \) is the standard deviation.
- **Using Statistical Software**: Libraries like Pandas in Python can compute skewness with functions like `DataFrame.skew()`.

### 3. How to Identify Left Skewed / Right Skewed?

You can identify the skewness of a distribution using the value of skewness:
- **Left-Skewed (Negative Skew)**: If skewness < 0, the left tail is longer. Most data points are concentrated on the right side.
- **Right-Skewed (Positive Skew)**: If skewness > 0, the right tail is longer. Most data points are concentrated on the left side.
- **Symmetrical**: If skewness  0, the data is fairly symmetric.

### 4. How to Find the Kurtosis Without a Graph?

Kurtosis measures the "tailedness" of the distribution, or the extremity of data points compared to a normal distribution:
- **Formula**:
  \[
  \text{Kurtosis} = \frac{n(n+1)}{(n-1)(n-2)(n-3)} \sum \left(\frac{x_i - \overline{x}}{s}\right)^4 - \frac{3(n-1)^2}{(n-2)(n-3)}
  \]
  where \( n \), \( x_i \), \( \overline{x} \), and \( s \) are as previously defined.
- **Using Statistical Software**: Libraries like Pandas can compute kurtosis with functions like `DataFrame.kurt()`.

### Summary:

- **Normal Distribution**: A symmetrical, bell-shaped distribution centered around the mean.
- **Skewness**: Measure of asymmetry. Can be found using a formula or statistical software.
- **Left/Right Skewness**: Determined by the sign of the skewness value.
- **Kurtosis**: Measure of tails' extremity. Can be found using a formula or statistical software.

These concepts are essential in statistics and can help in understanding the shape and characteristics of data distributions. If you have any more questions or need further explanations, feel free to ask! 
```
