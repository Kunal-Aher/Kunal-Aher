```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
```

```python
dfs=pd.read_csv(r"D:\kunal_cdac_blr_notes\DBDA SEP 2024\bmi.csv")
```

```python
dfs.drop(columns=['Unnamed: 6'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[10,6])
sns.boxplot(data=dfs)
```

```python
x=dfs[['Height', 'Weight ', 'Blood Pressure', 'Body Fat Percentag', 'BMI']]
y=dfs['Result']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=50)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
model=SVC(kernel="linear")
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
data=pd.DataFrame({"predicted":y_pred,"Actual":y_test})
data
```

```python
confu_matic=confusion_matrix(y_test,y_pred)
```

```python
confu_matic
```

```python
report=classification_report(y_test,y_pred)
print(report)
```
