Data Cleaning:

Handle missing values, duplicates, and any inconsistencies in the data.
Convert data types if necessary.

```python
duplicate_mask = super_df.duplicated()
duplicates = super_df[duplicate_mask]
duplicates
```

```python
# Remove duplicates
data_df= super_df.drop_duplicates()
```

```python
data_df.shape
```

```python
data_df.isnull().sum()
```

```python
data_df.info()
```

```python
city_counts = data_df['City'].value_counts()

highest_customers_city = city_counts.idxmax()
lowest_customers_city = city_counts.idxmin()

print(f"Highest Customers: {highest_customers_city}")
print(f"Lowest Customers: {lowest_customers_city}")
```

```python
# What is the distribution of sales per customer? (use histograms or boxplots)


# Assuming your DataFrame is named 'sales_df' and the sales column is 'sales'
plt.figure(figsize=(5,2))
plt.hist(super_df['Sales'], bins=15, edgecolor='black', color='skyblue')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Sales per Customer (Histogram)')
plt.show()
```

```python
country_sales_profits = data_df.groupby('Country').agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values(by='Sales', ascending=False)
print("Sales and profits by country:")
print(country_sales_profits)
```

```python

# Analyze sales and profits by city
city_sales_profits = data_df.groupby('City').agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values(by='Sales', ascending=False)
print("\nSales and profits by city:")
print(city_sales_profits)

# Analyze sales and profits by state
state_sales_profits = data_df.groupby('State').agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values(by='Sales', ascending=False)
print("\nSales and profits by state:")
print(state_sales_profits)
```

```python
# Check for regional differences in customer behavior
# For simplicity, let's consider 'Region' as a proxy for customer behavior
region_behavior = data_df.groupby('Region').agg({'Sales': 'mean', 'Quantity': 'mean', 'Discount': 'mean'})
print("\nRegional differences in customer behavior:")
print(region_behavior)
```

```python
# Visualize the sales by country
country_sales_profits['Sales'].plot(kind='bar', title='Sales by Country')
plt.show()

# Visualize the profits by country
country_sales_profits['Profit'].plot(kind='bar', title='Profits by Country', color='green')
plt.show()
```

```python
shipping_mode_sales = data_df.groupby('Ship Mode')['Sales'].sum()

# Create a bar plot to visualize the distribution
plt.figure(figsize=(8, 6))
plt.bar(shipping_mode_sales.index, shipping_mode_sales.values, color='skyblue', edgecolor='black')
plt.xlabel('Shipping Mode')
plt.ylabel('Total Sales')
plt.title('Distribution of Sales by Shipping Mode')
plt.show()
```

```python
Profit_by_segment=data_df.groupby("Segment")["Profit"].sum()
Profit_by_segment
```

```python
city_sales = data_df.groupby('City')['Sales'].sum()

# Sort cities by sales
sorted_cities_by_sales = city_sales.sort_values(ascending=False)

# Print the top cities
print("\nTop Cities (by Sales):")
print(sorted_cities_by_sales.head())
```

```python
# Assuming your DataFrame is named 'super_df'
region_sales_profit = data_df.groupby('Region')[['Sales', 'Profit']].sum()

# Create a bar plot to compare sales and profit by region
region_sales_profit.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Region')
plt.ylabel('Amount')
plt.title('Sales and Profit by Region')
plt.legend(['Sales', 'Profit'])
plt.show()
```

```python
# Assuming your DataFrame is named 'super_df' and the city column is 'City'
city_sales = super_df.groupby('City')['Sales'].sum()

# Sort cities by sales
sorted_cities_by_sales = city_sales.sort_values(ascending=False)

# Create a bar plot for top cities by sales
sorted_cities_by_sales.head(10).plot(kind='bar', figsize=(10, 6), color='skyblue', edgecolor='black')
plt.xlabel('City')
plt.ylabel('Sales Amount')
plt.title('Top Cities with High Sales')
plt.show()
```

```python
# Create a bar plot for bottom cities by sales
sorted_cities_by_sales.tail(10).plot(kind='bar', figsize=(10, 6), color='salmon', edgecolor='black')

plt.xlabel('City')
plt.ylabel('Sales Amount')
plt.title('Bottom Cities with Low Sales')
plt.show()
```
